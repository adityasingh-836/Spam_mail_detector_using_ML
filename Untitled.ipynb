```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler

# Load CSV file
dataset = pd.read_csv('Google_Stock_Price_Train.csv')

# Clean numeric columns (remove commas and convert to float)
for col in ['Open', 'High', 'Low', 'Close']:
    dataset[col] = dataset[col].astype(str).str.replace(',', '').astype(float)

# Clean Volume column
dataset['Volume'] = dataset['Volume'].astype(str).str.replace(',', '').astype(int)

# Convert Date column to datetime
dataset['Date'] = pd.to_datetime(dataset['Date'])

# Sort by date just in case
dataset = dataset.sort_values('Date')

# Feature and Target selection
features = ['Open', 'High', 'Low', 'Volume']
X = dataset[features].values
y = dataset['Close'].values

# Time-based split: 80% train, 20% test
split_index = int(0.8 * len(dataset))
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]

# Scale features
sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)

# Train Random Forest Regressor
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train_scaled, y_train)

# Predict
y_pred = rf.predict(X_test_scaled)

# Plot actual vs predicted
plt.figure(figsize=(10,6))
plt.plot(y_test, color='red', label='Actual Google Stock Price')
plt.plot(y_pred, color='blue', label='Predicted Stock Price (RF)')
plt.title('Google Stock Price Prediction using Random Forest')
plt.xlabel('Time Index')
plt.ylabel('Stock Price')
plt.legend()
plt.show()

# Evaluation
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f'Root Mean Squared Error (RMSE): {rmse:.2f}')
```
